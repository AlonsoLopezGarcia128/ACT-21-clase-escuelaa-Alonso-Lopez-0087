El texto describe la **abstracción de una clase** llamada **Escuela** bajo el modelo de Programación Orientada a Objetos (POO). En este esquema, se definen los **atributos**, que representan los datos o características de la institución (nombre, nivel, cantidad de alumnos y ubicación), y las **funciones** (o métodos), que establecen las acciones lógicas que el sistema puede ejecutar, tales como el registro de estudiantes, la gestión académica de notas y la planeación de actividades institucionales. Básicamente, funciona como un plano estructural para digitalizar el funcionamiento básico de un centro educativo.



![texto del vínculo](https://raw.githubusercontent.com/AlonsoLopezGarcia128/ACT-21-clase-escuelaa-Alonso-Lopez-0087/refs/heads/main/infoclaseescuela.png)

```python
# Definición de la Clase Escuela
class Escuela:
    # Método constructor para inicializar los atributos
    def __init__(self, nombre, nivelEducativo, numeroAlumnos, direccion):
        self.nombre = nombre
        self.nivelEducativo = nivelEducativo
        self.numeroAlumnos = numeroAlumnos
        self.direccion = direccion

    # Función 1: Matricular Alumno (Interactiva)
    def matricularAlumno(self):
        nombre_alumno = input(f"[{self.nombre}] Ingrese el nombre del alumno a matricular: ")
        print(f"📝 El alumno {nombre_alumno} ha sido inscrito exitosamente en el nivel {self.nivelEducativo}.")

    # Función 2: Asignar Calificación (Interactiva)
    def asignarCalificacion(self):
        materia = input(f"[{self.nombre}] ¿De qué materia desea asignar calificación?: ")
        nota = input(f"Ingrese la calificación para {materia}: ")
        print(f"📊 Calificación de {nota} registrada en el sistema de {self.nombre}.")

    # Función 3: Organizar Evento (Interactiva)
    def organizarEvento(self):
        evento = input(f"[{self.nombre}] Nombre del evento a organizar: ")
        fecha = input(f"Ingrese la fecha del evento: ")
        print(f"🎉 Confirmado: El evento '{evento}' se realizará el {fecha} en la dirección {self.direccion}.")

# --- Creación de Objetos (Instancias) ---

# Objeto 1: Escuela1
escuela1 = Escuela("Centro de Bachillerato Tecnologico Industrial y de Servicios 123", "bachillerato", 2000, "C Nueva Zelanda 128")

# Objeto 2: Escuela2
escuela2 = Escuela("Esc. Sec. Tecnica #55", "Secundaria", 1000, "Calle Zelanda 1502")

# --- Ejecución Interactiva ---

# Interacción con Escuela 1
print(f"--- Gestión de: {escuela1.nombre} ---")
escuela1.matricularAlumno()
escuela1.asignarCalificacion()
escuela1.organizarEvento()

print("\n" + "-"*90 + "\n")

# Interacción con Escuela 2
print(f"--- Gestión de: {escuela2.nombre} ---")
escuela2.matricularAlumno()
escuela2.asignarCalificacion()
escuela2.organizarEvento()

print("\n" + "-"*90 + "\n")

print("Creador de software Alonso Lopez NL = 31")
```
